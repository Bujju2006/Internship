```python
#Importing the required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
'''
1) Write a python program to display all the header tags from wikipedia.org and make data frame
'''
#https://en.wikipedia.org/wiki/Main_Page

page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup=BeautifulSoup(page.content)
headers=[]
for i in soup.find_all('span',class_='mw-headline'):
    headers.append(i.text)
#headers
df=pd.DataFrame({'wiki_headers':headers})
df
```

```python
'''
2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame
'''
page=requests.get('https://presidentofindia.nic.in/former-presidents')
soup=BeautifulSoup(page.content)

presidentList=[]
termsList=[]
details=[]
for i in soup.find_all('div',class_="desc-sec"):
    details.append(i.text)

for det in details:
    split_text = det.split('\n')
    #print(split_text)
    split_text = [item for item in split_text if item.strip()]
    presidentList.append(split_text[0])
    termsList.append(split_text[1])

df=pd.DataFrame({'Name':presidentList,'Term ofoffice':termsList})
df
```

```python
'''
 Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame
i) Headline
ii) Time
iii) News Link
--[time class="LatestNews-timestamp"]
--[class="LatestNews-headline"]
--[a href]
'''

page=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
#soup

times=[]
headlines=[]
newsLink=[]

for i in soup.find_all('a',class_='LatestNews-headline'):
    headlines.append(i.text)
    newsLink.append(i.get('href'))

for i in soup.find_all('time',class_='LatestNews-timestamp'):
    times.append(i.text)
#print(times) 

df=pd.DataFrame({'Headline':headlines,'Time':times,'News Link':newsLink})
df
```

```python
'''
Write a python program to scrape the details of most downloaded articles from AI in last 90
days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details and make data frame
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

--Author:
span [class="sc-1w3fpd7-0 dnCnAO"]
- link and name
[class="sc-5smygv-0 fIXTHm"]
 - Year
span
[class="sc-1thf9ly-2 dvggWt"]
'''
##https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup=BeautifulSoup(page.content)
#soup

titles=[]
authors=[]
titlesLink=[]
publishDate=[]

for i in soup.find_all('a',class_='sc-5smygv-0 fIXTHm'):
    titles.append(i.text)
    titlesLink.append(i.get('href'))

for i in soup.find_all('span',class_='sc-1w3fpd7-0 dnCnAO'):
    authors.append(i.text)
#print(times) 
for i in soup.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    publishDate.append(i.text)

df=pd.DataFrame({'Paper Title':titles,'Authors':authors,'Published Date':publishDate,'Paper URL':titlesLink})
df
```

```python
'''
Write a python program to scrape mentioned details from dineout.co.inand make data frame
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

'''
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(page.content)
resturants_det=[]

for i in soup.find_all('div',class_='restnt-info cursor'):
    resturants_det.append(i.text)
resturants_det
ellipsis=[]
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    ellipsis.append(i.text)

for det in ellipsis:
    split_text = det.split('|')
    #print(split_text)
    split_text = [item for item in split_text if item.strip()]
    cuisine.append(split_text[1])

resturants_loc=[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    resturants_loc.append(i.text)
resturants_loc

images=[]
for i in soup.find_all('img',class_='no-img'):
    images.append(i['data-src'])
images

ratings=[]
for i in soup.find_all('div',class_='restnt-rating rating-4'):
    ratings.append(i.text)
ratings

df=pd.DataFrame({'Restaurant name':resturants_det,'Cuisine':cuisine,'Location':resturants_loc,'Ratings':ratings,'resturants_image':images})
df
```
